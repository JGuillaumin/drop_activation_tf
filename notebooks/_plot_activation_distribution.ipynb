```python
import tensorflow as tf
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sess = tf.InteractiveSession()
```

# ReLU

```python
input_tf = tf.placeholder(shape=[1000, 1000], dtype=tf.float32)

is_training = tf.placeholder(shape=[], dtype=tf.bool)

output_tf = tf.nn.relu(input_tf)
```

```python

input_np = np.random.normal(loc=1.0, scale=0.96, size=[1000, 1000])

feed_dict = {input_tf: input_np,
             is_training: True}
output_np = sess.run(output_tf, feed_dict=feed_dict)


plt.figure(figsize=(14, 14))
plt.subplot(221)
plt.imshow(input_np, cmap='gray', vmin=input_np.min(), vmax=input_np.max())
plt.subplot(222)
plt.imshow(output_np, cmap='gray', vmin=input_np.min(), vmax=input_np.max())
plt.show()
```

```python
f, axes = plt.subplots(2, sharex=True, figsize=(11,11))
_  =axes[0].hist(input_np.flatten(), bins=255, density=True, alpha=0.5, color='blue')
axes[0].set_title('Input vs Output activation distribution of ReLU')
_ = axes[1].hist(output_np.flatten(), bins=255, density=True, alpha=0.5, color='green')
plt.show()
```
